```python
import subprocess, os, signal, time

process = None  # global reference

def start_comfyui_gpu(port=8288):
    global process
    print(f"üöÄ Starting ComfyUI GPU mode on port {port} ...")
    # Create a new process group so we can kill it later
    process = subprocess.Popen(
        [
            "python", "/opt/comfyui/main.py",
            "--listen", "0.0.0.0", "--port", str(port)
        ],
        preexec_fn=os.setsid  # ensures a new session group
    )
    print(f"‚úÖ ComfyUI started with PID {process.pid}")

def stop_comfyui():
    global process
    if process is None:
        print("‚ö†Ô∏è No ComfyUI process running.")
        return
    print(f"üõë Stopping ComfyUI (PID {process.pid}) ...")
    try:
        # Kill the entire process group (main.py + children)
        os.killpg(os.getpgid(process.pid), signal.SIGTERM)
        process.wait(timeout=10)
        print("‚úÖ ComfyUI stopped successfully.")
    except Exception as e:
        print(f"‚ö†Ô∏è Error stopping ComfyUI: {e}")

# Example run
start_comfyui_gpu()

```

```python
import requests, json

COMFY = "http://localhost:8288"   # inside-container address

def check_comfy_models(server=COMFY):
    print(f"üîé Checking ComfyUI at {server} ...")

    # 1Ô∏è‚É£ connectivity
    try:
        r = requests.get(f"{server}/system_stats", timeout=10)
        r.raise_for_status()
        print("‚úÖ ComfyUI server reachable.\n")
    except Exception as e:
        print(f"‚ùå Cannot reach ComfyUI: {e}")
        return

    # 2Ô∏è‚É£ fetch available models
    try:
        r = requests.get(f"{server}/object_info", timeout=10)
        r.raise_for_status()
        data = r.json()
    except Exception as e:
        print(f"‚ùå Failed to query /object_info: {e}")
        return

    # 3Ô∏è‚É£ extract known model lists
    def safe_get(path):
        d = data
        for k in path:
            d = d.get(k, {})
        if isinstance(d, dict) and "values" in d:
            return d["values"]
        if isinstance(d, dict):
            return list(d.keys())
        return d

    print("üì¶ Detected models:\n")

    print("üß† Diffusion models (/models/diffusion_models):")
    print("   ", safe_get(["UNETLoader", "input", "unet_name", "values"]))

    print("\nüé≠ VAE models (/models/vae):")
    print("   ", safe_get(["VAELoader", "input", "vae_name", "values"]))

    print("\nüí¨ CLIP / Text encoders (/models/text_encoders):")
    clip_names = safe_get(["DualCLIPLoader", "input", "clip_name1", "values"])
    print("   ", clip_names)

    print("\n‚úÖ If your workflow JSON uses any of these filenames, it should validate successfully.\n")

check_comfy_models()
```
