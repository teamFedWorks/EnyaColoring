```python
import cv2
import numpy as np
from tqdm import tqdm

def concat_videos_side_by_side(video_path1, video_path2, output_path):
    """
    Concatenate two videos side by side while keeping original aspect ratios.

    Args:
        video_path1 (str): Path to the first input video.
        video_path2 (str): Path to the second input video.
        output_path (str): Path to save the output concatenated video.
    """

    # Open both videos
    cap1 = cv2.VideoCapture(video_path1)
    cap2 = cv2.VideoCapture(video_path2)

    if not cap1.isOpened() or not cap2.isOpened():
        raise ValueError("❌ Could not open one of the input videos!")

    # Read FPS and frame count from first video
    fps1 = cap1.get(cv2.CAP_PROP_FPS)
    fps2 = cap2.get(cv2.CAP_PROP_FPS)
    fps = min(fps1, fps2) if fps1 > 0 and fps2 > 0 else 30

    frame_count = int(min(
        cap1.get(cv2.CAP_PROP_FRAME_COUNT),
        cap2.get(cv2.CAP_PROP_FRAME_COUNT)
    ))

    # Read first frames to get sizes
    ret1, frame1 = cap1.read()
    ret2, frame2 = cap2.read()
    if not (ret1 and ret2):
        raise ValueError("❌ Could not read initial frames from videos!")

    # Match heights and preserve aspect ratio
    h1, w1 = frame1.shape[:2]
    h2, w2 = frame2.shape[:2]
    target_height = min(h1, h2)

    def resize_keep_aspect(frame, target_height):
        h, w = frame.shape[:2]
        new_w = int(w * target_height / h)
        return cv2.resize(frame, (new_w, target_height))

    frame1 = resize_keep_aspect(frame1, target_height)
    frame2 = resize_keep_aspect(frame2, target_height)

    # Output video properties
    total_width = frame1.shape[1] + frame2.shape[1]
    output_size = (total_width, target_height)
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')

    out = cv2.VideoWriter(output_path, fourcc, fps, output_size)
    if not out.isOpened():
        raise RuntimeError("❌ Could not open VideoWriter. Check codec or permissions.")

    print(f"✅ Starting side-by-side video creation...")
    print(f"   Input 1: {video_path1}")
    print(f"   Input 2: {video_path2}")
    print(f"   Output : {output_path}")
    print(f"   Resolution: {output_size}, FPS: {fps}")

    pbar = tqdm(total=frame_count, desc="Merging frames")

    # Reset videos to frame 0
    cap1.set(cv2.CAP_PROP_POS_FRAMES, 0)
    cap2.set(cv2.CAP_PROP_POS_FRAMES, 0)

    while True:
        ret1, frame1 = cap1.read()
        ret2, frame2 = cap2.read()
        if not (ret1 and ret2):
            break

        frame1 = resize_keep_aspect(frame1, target_height)
        frame2 = resize_keep_aspect(frame2, target_height)

        # Concatenate horizontally
        combined = np.hstack((frame1, frame2))
        out.write(combined)
        pbar.update(1)

    pbar.close()
    cap1.release()
    cap2.release()
    out.release()

    print(f"✅ Done! Saved to: {output_path}")


# ---------- Example usage ----------
if __name__ == "__main__":
    video1 = "input_videos/t8.mp4"   # Replace with your left video path
    video2 = "results_realesrgan_1/t8_video_out.mp4"  # Replace with your right video path
    output = "results_realesrgan_1/t8_output_side_by_side.mp4"

    concat_videos_side_by_side(video1, video2, output)
```

```python
import subprocess

input_video = "output_videos_latest/TellacheeraKattukundievarikosamo_20251119012314_b9c3503b/final_without_post_process__b129636632ba/tellaCheera_portrait.mp4"     # <-- replace with your video name
output_video = "output_videos_latest/TellacheeraKattukundievarikosamo_20251119012314_b9c3503b/final_without_post_process__b129636632ba/tellaCheera_portrait_b&w.mp4"

# FFmpeg command:
# -vf "format=gray" → makes video black & white
# -c:v libx264 → encode video
# -crf 18 → high quality
# -preset medium → encoding speed (can change)
# -c:a copy → keep original audio intact
# FPS is automatically preserved unless changed explicitly
cmd = [
    "ffmpeg",
    "-i", input_video,
    "-vf", "format=gray",
    "-c:v", "libx264",
    "-crf", "18",
    "-preset", "medium",
    "-c:a", "copy",
    output_video
]

subprocess.run(cmd)
print("✔ Black & white video created:", output_video)
```

```python
import subprocess

top_video = "output_videos_latest/TellacheeraKattukundievarikosamo_20251119012314_b9c3503b/final_without_post_process__b129636632ba/tellaCheera_portrait.mp4"      # the video that will appear at the top
bottom_video = "output_videos_latest/TellacheeraKattukundievarikosamo_20251119012314_b9c3503b/final_without_post_process__b129636632ba/tellaCheera_portrait_b&w.mp4"   # the video that will appear at the bottom
output_video = "output_videos_latest/TellacheeraKattukundievarikosamo_20251119012314_b9c3503b/final_without_post_process__b129636632ba/tellaCheera_portrait_T&B.mp4"

# FFmpeg command:
# -filter_complex "vstack=inputs=2" → stacks videos vertically
# -c:v libx264 → video codec
# -crf 18 → high quality
# -preset medium → encoding speed
# -c:a copy → copies audio from the first input video
cmd = [
    "ffmpeg",
    "-i", top_video,
    "-i", bottom_video,
    "-filter_complex", "vstack=inputs=2",
    "-c:v", "libx264",
    "-crf", "18",
    "-preset", "medium",
    "-c:a", "copy",   # keeps audio from the top video
    output_video
]

subprocess.run(cmd)
print("✔ Portrait video created:", output_video)
```
